1. What is the difference between a function and a method in Python?

    Functions are blocks of reusable code that are called independently, while methods are functions defined within a class and can only be called on objects of that class.

2. Explain the concept of function arguments and parameters in Python.

    Function arguments are the values passed to a function when it is called, and parameters are the variables that define the function's input in its definition.

3. What are the different ways to define and call a function in Python?

    Functions can be defined using the `def` keyword or `lambda` keyword (for anonymous functions). They can be called by using the function name followed by parentheses.

4. What is the purpose of the `return` statement in a Python function?

    The `return` statement is used to send a result back to the caller from a function.

5. What are iterators in Python and how do they differ from iterables?

    Iterators are objects that implement the `__iter__()` and `__next__()` methods. Iterables are objects that can return an iterator, such as lists, tuples, and strings.

6. Explain the concept of generators in Python and how they are defined.
    
    Generators are special types of iterators that yield values one at a time using the `yield` keyword. They are defined using functions and `yield`.

7. What are the advantages of using generators over regular functions?
    
    Generators are memory efficient, as they generate values on the fly and avoid storing all items in memory like lists.

8. What is a lambda function in Python and when is it typically used?
    
    A lambda function is a small, anonymous function defined with the `lambda` keyword, used when a short, single-expression function is needed.

9. Explain the purpose and usage of the `map()` function in Python.
    
    The `map()` function applies a given function to each item in an iterable and returns an iterator with the results.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    
    The `map()` function applies a function to each item, `filter()` extracts items that meet a condition, and `reduce()` applies a function cumulatively to the items.

```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage
print(sum_of_even_numbers([1, 2, 3, 4, 5, 6]))  # Output: 12
```

```python
def reverse_string(s):
    return s[::-1]

# Example usage
print(reverse_string("hello"))  # Output: "olleh"
```

```python
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

# Example usage
print(square_numbers([1, 2, 3, 4]))  # Output: [1, 4, 9, 16]
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage
for num in range(1, 201):
    if is_prime(num):
        print(num)  # Prints all prime numbers from 1 to 200
```

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.current_term = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term < self.terms:
            self.current_term += 1
            self.a, self.b = self.b, self.a + self.b
            return self.a
        else:
            raise StopIteration

# Example usage
fibonacci = Fibonacci(10)
for num in fibonacci:
    print(num)  # Prints first 10 Fibonacci numbers
```

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_two(5):
    print(power)  # Output: 1, 2, 4, 8, 16, 32
```

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage (assuming 'sample.txt' is a file in the current directory)
for line in read_file_line_by_line('/content/drive/MyDrive/pwskills-DataScience_and_AI/Functions/sample.txt'):
    print(line)
```

```python
data = [(1, 'one'), (2, 'two'), (3, 'three')]
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)  # Output: [(1, 'one'), (3, 'three'), (2, 'two')]
```

```python
def celsius_to_fahrenheit(celsius):
    return list(map(lambda x: (x * 9/5) + 32, celsius))

# Example usage
temperatures = [0, 20, 37]
print(celsius_to_fahrenheit(temperatures))  # Output: [32.0, 68.0, 98.6]
```

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example usage
print(remove_vowels("Hello World"))  # Output: "Hll Wrld"
```

```python
# List of orders, where each sublist contains [Order Number, Book Title and Author, Quantity, Price per Item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Function to calculate total price with the specified adjustment
def calculate_total(order):
    order_number, _, quantity, price_per_item = order
    total_price = quantity * price_per_item
    # Add $10 if the total price is less than $100
    if total_price < 100:
        total_price += 10
    return (order_number, total_price)

# Using map with a lambda function to apply the calculation to each order
result = list(map(lambda order: calculate_total(order), orders))

# Print the result
print(result)
```
